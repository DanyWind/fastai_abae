```python
class SentenceEmbedder(nn.Module):
    def __init__(self,word_vectors,embedding_dim,freeze=True):
        super(SentenceEmbedder,self).__init__()
        
        # We initialize the word embeddings
        self.word_vectors = nn.Embedding(*word_vectors.shape)
        self.word_vectors.weight.data = torch.tensor(word_vectors)
        
        if freeze:
            self.word_vectors.weight.requires_grad = False
        
        self.M = nn.Linear(embedding_dim,embedding_dim,bias=False)
        
    def forward(self,x):
        # First we convert each word index to word embeddings
        x = self.word_vectors(x)
        
        # Then we pass it to the bilinear product
        y = x.mean(dim = 1)
        d = self.M(y)
        
        # We compute for each word of a sentence its product with My
        d = torch.einsum("ijk,ik -> ij",x,d)
        
        # We softmax the output
        a = F.softmax(d,dim = 1)
        
        # Finally we multiply and sum
        z = (a[:,:,None] * x).sum(dim = 1)
        
        # We store the z to later compute the loss
        self.z = z
        self.n = y
        
        return z
    
class SentenceAutoEncoder(nn.Module):
    def __init__(self,aspect_vectors,embedding_dim,n_aspect):
        super(SentenceAutoEncoder,self).__init__()
        
        self.W = nn.Linear(embedding_dim,n_aspect)
        
        self.T = nn.Linear(n_aspect,embedding_dim,bias=False)
        self.T.weight.data = torch.tensor(aspect_vectors)
        
    def forward(self,z):
        # First we project our sentence embedding to its aspects
        p = self.W(z)
        p = F.softmax(p,dim = 1)
        
        # Finally we reconstruct the sentence embedding from its aspects
        r = self.T(p)
        
        return r
    
    def predict(self,z):
        
        # First we project our sentence embedding to its aspects
        p = self.W(z)
        p = F.softmax(p,dim = 1)
        
        return p
```

```python
class ABAE(nn.Module):
    def __init__(self,word_vectors,aspect_vectors,embedding_dim,n_aspect):
        super(ABAE,self).__init__()
        
        self.embedder = SentenceEmbedder(word_vectors=word_vectors,embedding_dim=embedding_dim)
        self.autoencoder = SentenceAutoEncoder(aspect_vectors=aspect_vectors,embedding_dim=embedding_dim,
                                              n_aspect=n_aspect)
        
    def forward(self,x):
        z = self.embedder(x)
        r = self.autoencoder(z)
        
        return r  
    
    def predict(self,x):
        z = self.embedder(x)
        p = self.autoencoder.predict(z)
        
        return p
```

```python
model = ABAE(word_vectors=vectors,aspect_vectors=aspects,embedding_dim=d,n_aspect=k)
```

```python
def loss_func(r,y,abae,alpha=1,m=5):
    z = abae.embedder.z.detach()
    n = abae.embedder.n.detach()
    
    bs = z.shape[0]
    
    # We compute the Hinge loss
    positive = torch.einsum("ij,ij -> i",r,z)
    
    # We first create batches of negative samples
    negative_idx = []
    for i in range(bs):
        choice = np.delete(np.arange(bs),i)
        choice = np.random.choice(choice,m,replace=False)
        negative_idx.append(choice)

    negative_idx = np.stack(negative_idx)
    negative_idx = torch.tensor(negative_idx)
    
    # We broadcast first our batch then take the negative samples
    n = n.expand(bs,-1,-1)
    
    output = []
    for i in range(bs):
        v = n[i,negative_idx[i]]
        output.append(v)
    negative = torch.stack(output)
    
    # We compute the inner products
    negative = torch.einsum("ijk,ik -> ij",negative,r)
    
    # We compute the final loss
    J = (F.relu(1 + negative - positive[:,None])).sum(dim = 1).mean()
    
    # We compute the regularization loss
    T = abae.autoencoder.T.weight
    gram = torch.matmul(torch.t(T),T)
    
    k = gram.shape[0]
    I = torch.diag(torch.ones(k)).cuda()
    U = ((gram - I) ** 2).sum()
    
    loss = J + alpha * U
    
    return loss
```

```python
loss = partial(loss_func,abae=model)
```

```python
learn = Learner(data_lm,model,loss_func=loss)

learn.lr_find()
```

```python
def compute_k_closest_neighbors(vector,vectors,k=5):
    distances = np.sum((vectors - vector)**2,axis=1)
    distances = dict(zip(np.arange(len(distances)),distances))
    sorted_distances = sorted(distances, key = distances.get,reverse=True)
    output = sorted_distances[:k]
    
    return output
```
